**Histogram of Machine learning**

```
import matplotlib.pyplot as plt
import numpy as np
import random
```

**Generating Data**

```
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(15,40,(100))

print(ml_students_age)
print(py_students_age)
```

**histogram plt.hist()**

```
plt.hist(ml_students_age)
plt.show()
```

```
plt.hist(py_students_age)
plt.show()
```

**title/xlabel/ylabel**

```
plt.hist(ml_students_age)

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.show()
```

**bins**

```
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins)

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.show()
```

**rwidth**

```
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8)

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.show()
```

**histtype**

```
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8, histtype='step')

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.show()
```

**orientation**

```
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8, histtype='bar', orientation='horizontal', color='m')

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.show()
```

```
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8, histtype='bar', orientation='horizontal', color='m', label='ML Std')

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.legend()
plt.show()
```

**matplotlib style use**

```
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins, rwidth=0.8, histtype='bar', orientation='vertical', color='m', label='ML Std')
plt.hist(py_students_age,bins, rwidth=0.8, histtype='bar', orientation='vertical', color='y', label='PY Std')

plt.title('ML Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.legend()
plt.show()
```

**plot two histograms**

```
bins = [15,20,25,30,35,40,45]

plt.figure(figsize=(16,9))

plt.hist([ml_students_age,py_students_age],bins, rwidth=0.8, histtype='bar', orientation='vertical', color=['m','y'], label=['ML Std','PY Std'])

plt.title('ML & PY Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.legend()
plt.show()
```

**matplotlib grid**

```
from matplotlib import style
style.use('ggplot')
bins = [15,20,25,30,35,40,45]

plt.figure(figsize=(16,9))

plt.hist([ml_students_age,py_students_age],bins, rwidth=0.8, histtype='bar', orientation='vertical', color=['m','y'], label=['ML Std','PY Std'])

plt.title('ML & PY Students Age in Histogram')
plt.xlabel('Students Age Category')
plt.ylabel('No. of Students Age')
plt.legend()
plt.show()
```
