**Matplotlib Pie Chart**

```
import matplotlib.pyplot as plt
```

**Importing Dataset**

```
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']

class1_students = [45, 15, 35, 25, 30]
```

```
plt.pie([1]) 
plt.show()
```

**Plotting pie chart using real dataset**

```
plt.pie(class1_students, labels = classes)
plt.show()
```

**Colors**

```
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30]

exp=[0.03,0.03,0.03,0.03,0.03]
clr=['c','b','r','y','g']
plt.pie(class1_students, labels = classes,explode=exp, colors=clr)
plt.show()
```

**Auto Percentence**

```
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30]

exp=[0.03,0.03,0.03,0.03,0.03]
clr=['c','b','r','y','g']


plt.pie(class1_students, labels = classes,explode=exp, colors=clr, autopct='%0.2f%%')
plt.show()
```

**Radius**

**Text Properties**

```
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30]

exp=[0.03,0.03,0.03,0.03,0.03]
clr=['c','b','r','y','g']
textp={'fontsize':15}

plt.pie(class1_students, 
        labels = classes,
        explode=exp, 
        colors=clr, 
        autopct='%0.2f%%',
        shadow=True, 
        radius=2,
        startangle=90,
        counterclock=False,
        textprops=textp)
plt.show()
```

**Plotting pie chart using more parameters**

```
classes = ["Python", 'R', 'Machine Learning', 'Artificial Intelligence', 'Data Sciece']
class1_students = [45, 15, 35, 25, 30]
plt.figure(figsize = (3,2))


exp=[0.03,0.03,0.03,0.03,0.03]
clr=['c','b','r','y','g']
textp={'fontsize':15}
wedgeprops = {"linewidth": 4, 'width':1, "edgecolor":"k"}


plt.pie(class1_students, 
        labels = classes,
        explode=exp, 
        colors=clr, 
        autopct='%0.2f%%',
        shadow=True, 
        radius=1.1,
        startangle=270,
        counterclock=True,
        textprops=textp,
        pctdistance = 0.6,
        labeldistance = 1.4,
        wedgeprops = wedgeprops,
        center=(2,3),
        frame=True,
        rotatelabels=True)


plt.show()
```
