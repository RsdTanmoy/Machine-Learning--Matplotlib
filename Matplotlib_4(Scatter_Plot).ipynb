```
df=pd.read_csv('/content/gdrive/MyDrive/Matplotlib/googleplaystore.csv',nrows = 1000)
df.shape
```

```
x = df["Rating"]
y = df["Reviews"]

plt.scatter(x,y) # Draw scatter plot with parameter x and y
plt.show()
```

```
plt.scatter(x,y)
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

**color**

```
plt.scatter(x,y,c='r')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

**one scatter plot using matplotlib**

```
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r', marker='*', s=100, alpha=.5, linewidths=5,edgecolors='g')
plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Ratings')
plt.ylabel('Reviews')
plt.show()
```

**multiple scatter plot using matplotlib**

```
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r', marker='*', s=100, alpha=.5, linewidths=5,edgecolors='g')
plt.scatter(x,df['Installs'],c='y', marker='o', s=100, alpha=.5, linewidths=5,edgecolors='c')

plt.title('Google Play Store Apps Scatter Plot')
plt.xlabel('Ratings')
plt.ylabel('Installs & Reviews')
plt.show()
```
