**Bar chart using matplotlib**

```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

**Dataset**

```
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

**Plot Bar Chart**

```
plt.bar(classes, class1_students)
plt.show()
```

```
plt.barh(classes, class1_students)
plt.show()
```

**width**

```
plt.bar(classes, class1_students,width=0.5)
plt.show()
```

**align**

```
plt.bar(classes, class1_students,width=0.2, align='edge')
plt.show()
```

**color**

```
plt.bar(classes, class1_students,width=0.2, align='edge',color='y')
plt.show()
```

**edge color**

```
plt.bar(classes, class1_students,width=0.5, align='edge',color='y', edgecolor='m')
plt.show()
```

**line width**

```
plt.bar(classes, class1_students,width=0.5, align='edge',color='y', edgecolor='m', linewidth=5)
plt.show()
```

**alpha color**

```
plt.bar(classes, class1_students,width=0.5, align='edge',color='y', edgecolor='m', linewidth=5, alpha=0.5)
plt.show()
```

**line style**

```
plt.bar(classes, class1_students,width=0.5, align='edge',color='y', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--')
plt.show()
```

**label**

```
plt.bar(classes, class1_students,width=0.5, align='edge',color='y', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--', label='Class_1_Stduent')
plt.legend()
plt.show()
```

**Visible**

```
plt.bar(classes, class1_students,width=0.5, align='edge',color='y', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--', label='Class_1_Stduent', visible= False)
plt.legend()
plt.show()
```

```
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.bar(classes, class1_students,width=.6, align='edge',color='k', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--')
plt.show()
```

```

plt.figure(figsize=(16,9))
plt.barh(classes, class1_students, align='edge',color='k', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--') #del width
plt.show()
```

```
plt.figure(figsize=(16,9))
plt.bar(classes, class1_students,width=.6, align='center',color='k', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--')

plt.title('Bar Chart of Python Classes')
plt.xlabel('Classes')
plt.ylabel("No. of the Students")
plt.show()
```

```
plt.figure(figsize=(16,9))
plt.bar(classes, class1_students,width=.6, align='center',color='k', edgecolor='m', linewidth=5, alpha=0.5, linestyle='--')

plt.title('Bar Chart of Python Classes', fontsize=15)
plt.xlabel('Classes', fontsize=15)
plt.ylabel("No. of the Students", fontsize=15)
plt.show()
```

**plot two Bar chart using matplotlib**

```
plt.bar(classes, class1_students, color='b', label='Class_1 Students')
plt.bar(classes, class2_students, color='g', label='Class_2 Students')

plt.title('Bar Chart of Classes')
plt.xlabel('Classes')
plt.ylabel("No. of the Students")
plt.legend()
plt.show()
```

**Complete Bar Graph**

```
width=0.3
class_index=np.arange(len(classes))

plt.bar(class_index, class1_students,width, color='b', label='Class_1 Students')
plt.bar(class_index+width, class2_students,width, color='g', label='Class_2 Students')
plt.bar(class_index+width+width, class3_students,width, color='y', label='Class_3 Students')

plt.xticks(class_index+width, classes, rotation=20)
plt.title('Bar Chart of Classes')
plt.xlabel('Classes')
plt.ylabel("No. of the Students")
plt.legend()
plt.show()
```

```
width=0.3
class_index=np.arange(len(classes))

plt.barh(class_index, class1_students,width, color='b', label='Class_1 Students')
plt.barh(class_index+width, class2_students,width, color='g', label='Class_2 Students')
plt.barh(class_index+width+width, class3_students,width, color='y', label='Class_3 Students')

plt.yticks(class_index+width, classes, rotation=20)
plt.title('Bar Chart of Classes')
plt.ylabel('Classes')
plt.xlabel("No. of the Students")
plt.legend()
plt.show()
```
